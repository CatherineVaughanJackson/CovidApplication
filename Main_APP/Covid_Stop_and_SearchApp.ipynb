```python
import tkinter as tk
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, NavigationToolbar2Tk)
from matplotlib.figure import Figure
%matplotlib qt

import Geography_lists as Geo
import Date_range as ti
from PIL import ImageTk, Image
from tkinter import messagebox

from datetime import date
import requests
import json

import Window_1
import Window_2
import Window_3
import Window_4
```

```python
class App(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("Main Window Visulisation Interface")
        self.configure(background = "grey22")
        self.iconbitmap("coronavirus_orange_right_HPY_2.ico")
        
        self.create_buttons()
        self.create_title()
        self.image()
        
        
    def image(self):
        
        self.logo = tk.PhotoImage(master = self, file="covidgif.gif")
        label = tk.Label(self, image=self.logo)
        label.grid(column = 1, row= 2, rowspan = 3, pady = 20, sticky = "nesw")
        
    def create_title(self):
        
        #TITLE
        
        border_color = tk.Frame(self,background="gold2")
        main_headerlabel = tk.Label(border_color, text = "Stop and Search and Covid-19 Data Visulisation Interface",
                               bg = "grey30", relief = "solid", fg = "white", font=("Arial", 24, "bold"))
        
        border_color.grid( column= 0, row = 0,columnspan=4, padx =20, pady= 10)
        main_headerlabel.grid( column= 0, row = 0,columnspan=4, padx =3, pady= 3, sticky = "nesw")
        
        #STOP AND SEARCH HEADER
        
        left_border_color = tk.Frame(self,background="gold2")
        left_headerlabel = tk.Label(left_border_color, text = "Stop and Search Data",
                               bg = "grey30", relief = "solid", fg = "white", font=("Arial", 24, "bold"))
        
        left_border_color.grid( column= 0, row = 1, padx =20, pady= 10, sticky = "nesw")
        left_headerlabel.grid( column= 0, row = 1, padx =3, pady= 3, sticky = "nesw")
        
        #COVID VISULISATION HEADER
        
        right_border_color = tk.Frame(self,background="gold2")
        right_headerlabel = tk.Label(right_border_color, text = "Covid-19 Statistics",
                               bg = "grey30", relief = "solid", fg = "white", font=("Arial", 24, "bold"))
        
        right_border_color.grid( column= 3, row = 1, padx =20, pady= 10)
        right_headerlabel.grid( column=3 , row = 1,  padx =3, pady= 3, sticky = "nesw")
    
    def create_buttons(self):
        
        #FORMATTING
        
        formating_colour = {"bg" : "grey32","relief" : "raised", "fg": "white", "font":("Arial", 15)}
        formating_position = { "padx":40, "pady" :10, "sticky" :"nesw"}
        
        #COVID VISULISATION BUTTONS
        
        total_cases_button =tk.Button(self,text='Compare Number of Covid cases by area \n Cumlative and Daily',command=self.open_window_1, **formating_colour)
        total_cases_button.grid(column=3, row=2, **formating_position)
        
        area_change_button = tk.Button(self,text='Areas with the higest confirmed cases by time',command=self.open_window_2, **formating_colour)
        area_change_button.grid(column=3, row=3, **formating_position)
        
        country_cases_button =tk.Button(self,text='Country Total Number of cases over time',command=self.open_window_3, **formating_colour)
        country_cases_button.grid(column=3, row=4, **formating_position)
        

        #STOP AND SEARCH BUTTON
        
        sas_period =tk.Button(self,text='Stop and Search in a period\n by police force',command=self.open_window_4, **formating_colour)
        sas_period.grid(column=0, row=2, **formating_position)
        
        #MAIN QUIT BUTTON
        main_quit = tk.Button(self, text="QUIT", command = self.destroy, bg= "red3",fg= "black", relief = "raised", font=("Arial", 20) )
        main_quit.grid(column = 1, row = 5, pady=20, sticky = "nesw")
        
      
    def open_window_1(self):
        window = Window_1.Window1(self)
        window.grab_set()
        
    def open_window_2(self):
        window = Window_2.Window2(self)
        window.grab_set()
        
    def open_window_3(self):
        window = Window_3.Window3(self)
        window.grab_set()
        
    def open_window_4(self):
        window = Window_4.Window4(self)
        window.grab_set()

if __name__ == "__main__":
    app = App()
    app.mainloop()
```
